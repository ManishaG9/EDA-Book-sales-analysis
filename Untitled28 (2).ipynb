```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Books_Data_Clean.csv")
```

```python
# to see the first 5 rows of dataset
data.head(5)
```

```python
# here we get numerical summaries of data columns
data.describe()
```

```python
# to fix the negative number of publishing year - 560.000

data= data[data["Publishing Year"] > 1900]
```

```python
# number is change from -560.000 to 1901.000
data.describe()
```

```python
# to drop value of bookname = 0 
data.isna().sum()
```

```python
data.dropna(subset = "Book Name", inplace = True)
```

```python
# to see if we have any duplicated values in the datasets?

# so we don't have any duplicate values on table

data.duplicated().sum()
```

```python
# to check whether we have unique values on the table?

# as we can see we have unique values on 

data.nunique()
```

```python
# visualizing data through histrogram

# In this code it show's visualizes publishing year of books. on 2020 more books were published as compared to year 1900

plt.hist(data["Publishing Year"])
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title("DISTRIBUTION OF PUBLISHING YEAR")
plt.show()
```

```python

# In this code the bar charts show's the most highest genre was for fiction and lowest was for children

data["genre"].value_counts().plot(kind="bar")
plt.xlabel("Genre")
plt.ylabel("Number of books")
plt.title("NUMBER OF BOOKS IN EACH GENRE")
plt.show()
```

```python
type (data.groupby("Author")["Book_average_rating"].mean().reset_index())
```

```python
data.groupby("Author")["Book_average_rating"].mean().sort_values(ascending = False)
```

```python
sns.boxplot(x="genre", y= "Book_ratings_count", data = data)
plt.xlabel("Genre")
plt.ylabel("Book Ratings Count")
plt.title("Box plot of book ratings Count for Each Genre")
```

```python
language_counts = data["language_code"].value_counts()
```

```python
plt.pie(language_counts, labels= language_counts.index, startangle= 90, autopct = "%1.1f%%")
plt.title=("Language Distribution of Books")
plt.show()
```

```python
plt.scatter(data["sale price"], data["units sold"])
plt.xlabel("Sale price")
plt.ylabel("Units sold")
plt.title("Scatter Plot of Sale price vs Units sold")
plt.show()
```

```python
# Bar chart for total gross sales

total_gross_sales_by_author = data.groupby("Author")["gross sales"].sum()
```

```python
total_gross_sales_by_author.sort_values(ascending = False).head(10).plot(kind = "bar")
plt.xlabel=("Author")
plt.ylabel=("Total Gross sales")
plt.xtitle=("Total Gross Sales By  Each Author")
plt.show()
```

```python
sns.boxplot(x="Author_Rating", y = "units sold", data = data)
plt.xlabel=("Author Rating")
plt.ylabel= ("Units Sold")
plt.title("Box Plot of Units Sold for Each  Author Rating")
plt.show()
```

```python
data.groupby("Publishing Year")["units sold"].sum().plot(kind = "line", marker ="o")
plt.xlabel= ("Publishing Year")
plt.ylabel= ("Units Sold")
plt.title("Units Sold As Per Each Publishing Year")
plt.show()
```

```python
# key insights 
# Data Cleaning: Fixed publishing year outliers (e.g., -560  filtered after 1900) and removed missing book names.

# Publishing Trends: Most books were published between 20002016, with 2012 being a peak year.

# Genres: Fiction was the most published and popular genre; childrens books were least represented.

#Authors:

#Top-rated authors included Bill Watterson, J.R.R. Tolkien, George R.R. Martin.

# Some authors (like Chetan Bhagat, Sue Monk Kidd) had lower ratings.

# Sales & Revenue:

# Certain authors generated the highest gross sales.

# Units sold varied with author reputation and average rating.

# Languages: English dominated (ENG, EN-US, EN-GB), while non-English (fre, spa, ara) had minimal representation.
```
